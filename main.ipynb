```python
import pandas as pd
df = pd.read_csv("songs_with_audio_features.csv")
```

```python
#only have the last line of the df in a new df
ndf = df.tail(10)
ndf
```

```python
ndf = ndf.drop(['spotify_id', 'Artist', 'Album', 'Song Name', 'like'], axis=1)
```

```python
ndf.info()
```

```python
#predict the like for the ndf using the model saved in rf_model.sav
import pickle
loaded_model = pickle.load(open('rf_model.sav', 'rb'))
result = loaded_model.predict(ndf)
print(result)
```
